```python
import numpy as np
import pandas as pd 
import seaborn as sns
import math 
from scipy.special import factorial

train_x = np.load('D:/.ipynb_checkpoints/train_features.npy')
train_label = np.load('D:/.ipynb_checkpoints/train_labels.npy')
#print(Data)
#print(Data[0])
print(train_x.shape)
train_label.reshape(-1 ,1)
print(train_label.shape)


def compute_probs(X, y):
    
    # Find the number of classes
    num_classes = int(np.max(y)) + 1

    # Compute the probabilities
    pc = []
    p_x_given_c = []
    
    for c in range(num_classes):
        idx = []
        idx = y[:,0] == c
#         for i in range(4459):
#             if y[i]==c:
#                 idx.append(1)
#             else:
#                 idx.append(0)
        #print(idx)
        
        # Compute the prior
        pc.append(np.mean(idx))
        #print(idx)
        print("value of pc")
        print(pc)

        # Compute the conditional
        p_x_given_c.append(X[idx, :].mean(axis=0).reshape((-1, 1)))
#         m_all_y  = []
#         for j in range(954):
#             m_j_y = []
#             for k in range(4459):
#                 print(idx[1])
#                 if y[k]==c:
#                 m_j_y.append(X[k,j])    
#             m_all_y.append(np.mean(m_j_y))
#         p_x_given_c.append(m_all_y)
#         print("prob of x given c" , len(p_x_given_c))
        #print(p_x_given_c)
    return p_x_given_c, pc


def predict(X, p_x_given_c, pc):
    
    # Counts
    num_sentences = X.shape[0]
    num_classes = len(p_x_given_c)

    posterior = np.zeros(num_sentences)
    # Compute predi0ctions
#     for s in range(num_sentences):
#         (class_1, class_2) = (1,1)
#         for k in range(X.shape[1]):
#             n = X[s][k]
#             class_1 = class_1*1/math.factorial(n)*np.exp(-p_x_given_c[0][k])*math.pow(p_x_given_c[0][k], n)
#             class_2 = class_2*1/math.factorial(n)*np.exp(-p_x_given_c[1][k])*math.pow(p_x_given_c[1][k], n)
#         class_1 = class_1*pc[0]
#         class_2 = class_2*pc[1]
#         if(class_1>class_2):
#             posterior[s] = 0
#         else:
#             posterior[s] = 1
   
#     return posterior
    log_posterior = np.zeros((num_sentences, num_classes))
    for c in range(num_classes):
        m_j_y = p_x_given_c[c].T.repeat(num_sentences, axis=0)
        log_prob = X * np.log(1e-10 + m_j_y) - (1e-10 + m_j_y) - np.log(factorial(X))
        sum_log_prob = (np.sum(log_prob , axis=1)) #np.exp(np.log(1e-12 + probs).sum(axis=1)
        log_posterior[:, c] = sum_log_prob + np.log(pc[c])

    # Assign the labels
    preds = np.argmax(log_posterior, axis=1).reshape((-1, 1))

    return preds

p_x_given_c, pc = compute_probs(train_x, train_label.reshape(-1,1))
#print("log of prior")
#print(np.log(pc[0]))
#print(np.log(pc[1]))
#print("given that",p_x_given_c[0])
#print("given that 2 " , p_x_given_c[1])
                  
pred_train = predict(train_x, p_x_given_c, pc)
print('Training accuracy:', (pred_train.reshape((-1,)) == train_label.reshape((-1,))).mean())

test_x = np.load('D:/.ipynb_checkpoints/test_features.npy')
test_label = np.load('D:/.ipynb_checkpoints/test_labels.npy')
#print(Data)
#print(Data[0])
print(test_x.shape)
test_label.reshape(-1 ,1)
print(test_label.shape)

pred_train = predict(test_x, p_x_given_c, pc)
print('Testing accuracy:', (pred_train.reshape((-1,)) == test_label.reshape((-1,))).mean())

```
